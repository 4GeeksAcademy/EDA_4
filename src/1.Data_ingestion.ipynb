```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/4GeeksAcademy/regularized-linear-regression-project-tutorial/main/demographic_health_data.csv"

df = pd.read_csv(url)
df.to_csv("/workspaces/EDA_4/data/raw/demographic_health_data.csv")

print(df.columns)
```

```python
valores_nulos = df.isnull().sum()
valores_nulos.to_csv("/workspaces/EDA_4/data/interim/nulls.csv")
```

```python
df.rename(columns={
    "Active Physicians per 100000 Population 2018 (AAMC)" : "Physicians/10000", 
    "Active Primary Care Physicians per 100000 Population 2018 (AAMC)" : "Primary Care Physicians/10000",
    "Active Patient Care Primary Care Physicians per 100000 Population 2018 (AAMC)" : "Patient Care Primary Care Physicians/10000",
    "Active General Surgeons per 100000 Population 2018 (AAMC)" : "General Surgeons/10000",
    "Active Patient Care General Surgeons per 100000 Population 2018 (AAMC)" : "Patient Care General Surgeons/10000"
    }, inplace= True)
```

```python
columns = [
        "Physicians/10000", 
        "Primary Care Physicians/10000", 
        "Patient Care Primary Care Physicians/10000",
        "General Surgeons/10000",
        "Patient Care General Surgeons/10000",
        "Median_Household_Income_2018", "Percent of Population Aged 60+", 
        "STATE_NAME", 
        "anycondition_prevalence", 
        "Obesity_prevalence",  
        "COPD_prevalence", 
        "diabetes_prevalence", 
        "CKD_prevalence", 
        "Heart disease_prevalence"
    ]
```

```python
total_data = df[columns]
total_data
```

```python
total_data_gropued = total_data.groupby("STATE_NAME").mean()
total_data_gropued = total_data_gropued.reset_index()
total_data_gropued
```

```python
total_data_gropued.to_csv('/workspaces/EDA_4/data/processed/conditions_per_state.csv', index= False)
```
